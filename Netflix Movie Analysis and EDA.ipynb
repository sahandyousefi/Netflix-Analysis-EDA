# EDA 

```python
# I'm going to find number of movies produced by countries

data.production_countries.value_counts()
```

```python
top_10=data.production_countries.value_counts().head(10)
```

```python
top_10.plot(kind='bar', width=0.8, figsize=(9,6), color='red')
plt.title('Top 10')
plt.xlabel('Countries')
plt.ylabel('count')
plt.show()
```

```python
top_10.plot(kind='line', figsize=(9,6), color='red')
plt.title('Top 10')
plt.xlabel('Countries')
plt.ylabel('count')
plt.show()
```

```python
# Analyzing number of movies per year

c_year=data.release_year.value_counts()
c_year
```

```python
c_year.plot(kind='line', figsize=(9,6), color='red')
plt.title('number of movies per year')
plt.xlabel('years')
plt.ylabel('count')
plt.show()
```

```python
c_year.plot(kind='bar', figsize=(9,6), color='b')
plt.title('number of movies per year')
plt.xlabel('years')
plt.ylabel('count')
plt.show()
```

```python
# Top 10 movies with highest rating based on tmdb score and popularity
top_10_based_on_tmdb_rating = data.sort_values(['tmdb_score', 'tmdb_popularity'], ascending= False)[['title','tmdb_score','tmdb_popularity', 'type']].head(10)
top_10_based_on_tmdb_rating
```

```python
# Rank based on TMDB rates

top_10_based_on_tmdb_rating.plot(kind='barh', x = 'title', y = 'tmdb_popularity', figsize=(11, 8), color = 'green')
plt.title('Top 10 based on tmdb votes')
plt.xlabel('tmdb')
plt.ylabel('Name')
plt.show()
```

```python
# top directors and actors
# merge the datasets
data = data.merge(credit, how = 'outer', on= 'id')
```

```python
data.head()
```

```python
# filter only directors
director = data[data['role'] == 'DIRECTOR']
director.head()
```

```python
director=data[data['role']=='DIRECTOR']
```

```python
director.head(20)
```

```python
#Director ranking based on tmdb score and popularity

top_director = director.sort_values(['tmdb_score','tmdb_popularity'], ascending = False)[['name', 'tmdb_score','tmdb_popularity']].head(30)
top_director
```

```python
top_director.plot(kind= 'barh', y = 'tmdb_popularity', x = 'name', figsize= (9,6), color= 'red')
plt.xlabel('tmdb')
plt.ylabel('Director name')
plt.title('Top directors based on tmdb popularity')
plt.show()
```
